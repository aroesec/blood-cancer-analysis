Load Kaggle Data

```python
import kagglehub
import os
import pandas as pd
from sklearn.model_selection import train_test_split
import cv2


path = kagglehub.dataset_download("mahdinavaei/blood-cancer")

# The dataset contains images, not CSV files
dataset_dir = os.path.join(path, "bloodcells_dataset")

# List all cell types (directories)
cell_types = [d for d in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, d))]
print("Cell types in the dataset:")
for cell_type in cell_types:
    cell_type_path = os.path.join(dataset_dir, cell_type)
    num_images = len([f for f in os.listdir(cell_type_path) if f.endswith('.jpg')])
    print(f"  {cell_type}: {num_images} images")

# If you need to work with the data as a DataFrame, you can create one
data = []
for cell_type in cell_types:
    cell_type_path = os.path.join(dataset_dir, cell_type)
    images = [f for f in os.listdir(cell_type_path) if f.endswith('.jpg')]
    for image in images:
        data.append({
            'cell_type': cell_type,
            'image_name': image,
            'image_path': os.path.join(cell_type_path, image)
        })

df = pd.DataFrame(data)
print(f"\nTotal images: {len(df)}")
df.head()
```

```python
train_df, dummy_df = train_test_split(df,  train_size= 0.8, shuffle= True, random_state= 123)

valid_df, test_df = train_test_split(dummy_df,  train_size= 0.6, shuffle= True, random_state= 123)
```
